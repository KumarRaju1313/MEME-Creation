```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
```

```python
Img = Image.open("meme_5.jpg")
Font = ImageFont.truetype("arial.ttf", 40)

Top_text = "Training a Model for 10 Hours\n\n Getting 55% Accuracy"
Bottom_text = "Hyperparameter Tuning \n\n Getting 95% Accuracy"

Draw = ImageDraw.Draw(Img)
Top_text_position = (650, 30)
Draw.text(Top_text_position, Top_text, font = Font, fill = "black")

Bottom_text_position = (650, Img.height - 250)
Draw.text(Bottom_text_position, Bottom_text, font = Font, fill = "black")
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(Img)
plt.axis('off') 
plt.show()
```
